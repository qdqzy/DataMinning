#### 基本平稳
#### 自回归模型 描述当前值与历时值之前的关系，用变量自身的历史数据对自身进行预测，满足的基本平稳了
## ARIMA（差分自回归移动平均模型 ）指定 p，q，d 三个参数
### AR是自回归， p为自回归项，MA为移动平均 q为移动平均系数 d为时间序列为平稳时所作的差分次数
### 原理：将非平稳时间序列转化为平稳时间序列然后将因变量仅对他的滞后值以及随机误差项的现值和滞后值进行回归所建立的模型

### 自相关函数 ACF:反应同一序列在不同时序的取值之间的相关性 [-1,1] 置信区间
### 偏自相关函数PACF

```python
#计算acf pcf 
import statsmodels.api as sm

def tsplot(y, lags = None, title = "", figsize = (14, 8)):
    
    fig = plt.figure(figsize=figsize)
    layout = (2, 2)
    ts_ax =plt.subplot2grid(layout, (0, 0))
    hist_ax = plt.subplot2grid(layout, (0, 1))
    acf_ax = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (0, 1))
    
    y.plot(ax = ts_ax)
   # ts_ax.set_titile(title)
   # hist_ax.set_titile('Historygram')
    sm.graphics.tsa.plot_acf(y, lags = lags, ax = acf_ax)
    sm.graphics.tsa.plot_pacf(y, lags = lags, ax = pacf_ax)
    [ax.set_xlim(0) for ax in [acf_ax,pacf_ax]]
    sns.despine()
    plt.tight_layout()
    return ts_ax, acf_ax, pacf_ax
    
```

```python
tsplot(weather_week_diff, title='First order difference',lags = 10)
```

```python
fig = plt.figure(figsize=(12,8))
ax1=fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(weather_week_diff,lags=20,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(weather_week_diff,lags=20,ax=ax2)
```
